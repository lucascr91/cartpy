```python
from cartpy import *
import matplotlib.pyplot as plt
```

```python
data1872=Year(1872)
```

```python
data1920=Year(1991)
```

```python
sf1=data1872.get_geodata(state='SP',county=3548906)
```

```python
sf2=data1920.get_geodata(state='SP',county='Sao Carlos')
```

```python
fig, ax = plt.subplots(figsize=(20,10))

ax.set_aspect('equal')

sf1.plot(ax=ax,color='w', edgecolor='k', label=str(data1872.year))
sf2.plot(ax=ax,color='w', edgecolor='r', label=str(data1920.year))

ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.05),
          fancybox=True, shadow=True, ncol=5)
plt.show()
```
