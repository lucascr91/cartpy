```python
from cartpy import *
import matplotlib.pyplot as plt
from descartes import PolygonPatch
BLUE = '#6699cc'
```

```python
data1872=Year(1872)
```

```python
data1920=Year(1991)
```

```python
jf=Municipio('Juiz de Fora')
jf.get_code(state='MG',year=1991)
```

```python
sf1=data1872.get_geodata(state='SP',county=3548906)
sf2=data1920.get_geodata(state='SP',county='Sao Carlos')
```

```python
def compare(years,state,county,fillcolor1='w',fillcolor2='r', edgecolor1='k', edgecolor2='r'):
    if (isinstance(years,list)) & (len(years)==2):
        sf1=Year(years[0]).get_geodata(state=state,county=county)
        sf2=Year(years[1]).get_geodata(state=state,county=county)
        fig, ax = plt.subplots(figsize=(20,10))
        poly1= sf1.reset_index()['geometry'][0]
        ax.add_patch(PolygonPatch(poly1, fc=fillcolor1, ec=edgecolor1, alpha=1, zorder=2,label=str(years[0])))
        ax.axis('scaled')
        poly2= sf2.reset_index()['geometry'][0]
        ax.add_patch(PolygonPatch(poly2, fc=fillcolor2,ec=edgecolor2, alpha=0.3, zorder=2, label=str(years[1])))
        ax.axis('scaled')
        ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.05),
                  fancybox=True, shadow=True, ncol=2)
        plt.show()
    else:
        raise Exception("years must be a list with 2 elements")
```

```python
compare([1872,1991],state='MG',county=3136702)
```
